```python
from math import sin, cos, sqrt, atan2, radians
from sklearn.neighbors import DistanceMetric
```

```python
def haversine_vectorize(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    newlon = lon2 - lon1
    newlat = lat2 - lat1

    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2

    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    miles = 3958 * dist #6367 for distance in KM for miles use 3958
    return miles
```

```python
bike20_df['distance'] = haversine_vectorize(bike20_df['start station latitude'],bike20_df['start station longitude'], bike20_df['end station latitude'],
                                           bike20_df['end station longitude'])
bike20_df.head()
```
